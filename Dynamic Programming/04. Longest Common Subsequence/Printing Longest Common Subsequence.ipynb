# [**Longest Common Subsequence**](https://leetcode.com/problems/longest-common-subsequence/)

```python
s:str= "ab"
''.join(sorted(s, reverse=True))
```

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        s1= text1
        m= len(s1)

        s2= text2
        n= len(s2)


        t= [[0 for i in range(m+1)] for j in range(n+1)]

        for i in range(1, n+1):
            for j in range(1, m+1):
                if s2[i-1]==s1[j-1]:
                    t[i][j]= 1+ t[i-1][j-1]
                else:
                    t[i][j]= max(
                        t[i-1][j],
                        t[i][j-1]
                    )

        i=n
        j=m
        ans= ""
        while i!=0 and j!=0:
            if s2[i-1]==s1[j-1]:
                ans+=s2[i-1]
                i-=1
                j-=1
            else:
                if t[i-1][j]>t[i][j-1]:
                    i-=1
                else:
                    j-=1

        ans= ans[::-1]
        print(ans)
        return t[-1][-1]
    
s= Solution()
```
