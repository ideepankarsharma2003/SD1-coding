# [**1312. Minimum Insertion Steps to Make a String Palindrome**](https://leetcode.com/problems/minimum-insertion-steps-to-make-a-string-palindrome/description/)<br>



```python
class Solution:
    def minInsertions(self, s: str) -> int:

        n= len(s)
        s1= s
        s2= s[::-1] # reverse of s

        # initialization
        t= [[0 for i in range(n+1)] for j in range(n+1)]

        # LCS
        for i in range(1, n+1):
            for j in range(1, n+1):
                if s1[i-1]==s2[j-1]:
                    t[i][j]= 1+ t[i-1][j-1]
                else:
                    t[i][j]= max(
                        t[i-1][j],
                        t[i][j-1]
                    )

        '''
        #insertions = #deletions
        '''
        lps= t[-1][-1]

        return n-lps
        
```
