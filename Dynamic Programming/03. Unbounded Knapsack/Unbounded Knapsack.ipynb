# [**Unbounded Knapsack**](https://practice.geeksforgeeks.org/problems/knapsack-with-duplicate-items4201/1)

```python

class Solution:
    def knapSack(self, N, W, val, wt):
        t= [[-1 for  i in range(W+1)] for j in range(N+1)]
        # i ---> n
        # j ---> w
        for i in range(N+1):
            for j in range(W+1):
                if j==0 or i==0:
                    t[i][j]=0
                
        for i in range(1, N+1):
            for j in range(1, W+1):
                # processed
                if wt[i-1]>j:
                    t[i][j]=t[i-1][j]

                # have choice
                else:
                    t[i][j]= max(
                        t[i-1][j], # not choosing
                        val[i-1]+   t[i][j-wt[i-1]] # unprocessed
                    )

        for i in t:
            print(i)
        return t[-1][-1]
    

s= Solution()

```

```python
n,w= 2, 3
val= [1, 1]
wt= [2, 1]

s.knapSack(n,w,val, wt)
```
