# [Sort012](https://practice.geeksforgeeks.org/problems/sort-an-array-of-0s-1s-and-2s4231/1)

```python
#User function Template for python3

class Solution:
    def sort012(self,arr,n):
        # code here
        left= 0
        right= n-1
        
        mid= 0
        
        while mid<=right:
            print(f"{left, mid, right}")
            if arr[mid]==0:
                arr[left], arr[mid]= arr[mid], arr[left]
                print("left\t", arr)
                left+=1; mid+=1;
                
            elif arr[mid]==1:
                mid+=1
            
            else:
                arr[right], arr[mid]= arr[mid], arr[right]
                print("right\t", arr)
                right-=1


s= Solution()
```

```python
n= 5
arr= [0, 2, 1, 2, 0]
s.sort012(arr, n)

arr
```

```python
class Solution:
    def findMaxSum(self,arr, n):
        # code here
        if n==0:
            return 0
        if n==1:
            return max(arr[0], 0)
        # print(n, arr[n-1])
        return max(
            arr[n-1]+ self.findMaxSum(arr, n-2),
            self.findMaxSum(arr, n-1)
		    )
    

s= Solution()
```

```python
N = 6
Arr = [5, 5, 10, 100, 10, 5]
arr= list(Arr)
s.findMaxSum(arr, N)
```
