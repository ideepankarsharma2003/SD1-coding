# [**Equal Sum Partition**](https://leetcode.com/problems/partition-equal-subset-sum/)

```python
from typing import List
class Solution:
    def subset_sum(self, S, arr, n):
        t= [[-1 for i in range(S+1)] for j in range(n+1)]
        for i in range(n+1):
            for j in range(S+1):
                if j==0:
                    # sum is zero
                    t[i][j]=True
                if i==0 and j>0:
                    # arr is empty and sum is not zero
                    t[i][j]=False

        print(t)
        for i in range(1, n+1):
            for j in range(1, S+1):
                if arr[i-1]>j:
                    t[i][j]= t[i-1][j]

                else:
                    t[i][j]= t[i-1][j] or t[i-1][j-arr[i-1]]  

        print(t)
        return t[-1][-1]           
                





    def canPartition(self, nums: List[int]) -> bool:
        S= sum(nums)
        if S%2!=0:
            return False
        return self.subset_sum(S//2, nums, len(nums))
        
s= Solution()
        
```

```python
nums= [1,5,11,5]
s.canPartition(nums)
```
