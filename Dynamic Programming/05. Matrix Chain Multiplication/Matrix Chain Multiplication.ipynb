# [**Matrix Chain Multiplication**](https://www.geeksforgeeks.org/problems/matrix-chain-multiplication0303/1)






# Recursive

```python
class Solution:
    def solve(self, arr, i, j):
        if i >= j:
            return 0

        ans = 10000000000000000

        for k in range(i, j):
            temp_ans = (
                self.solve(arr, i, k) +
                self.solve(arr, k+1, j) +
                arr[i-1]*arr[k]*arr[j]
            )
            # print(ans, i, j)

            ans = min(ans, temp_ans)

        return ans

    def matrixMultiplication(self, N, arr):
        # code here
        i = 1
        j = N-1

        return self.solve(arr, i, j)
```
