# [**Minimum number of deletions and insertions**](https://practice.geeksforgeeks.org/problems/minimum-number-of-deletions-and-insertions0209/1)<br>


```python
#User function Template for python3
class Solution:
	def minOperations(self, s1, s2):
		# code here
		n= len(s2)
		m= len(s1)
		
		# initialization
		t= [[0 for i in range(m+1)] for j in range(n+1)]
		
		# lcs
		for i in range(1, n+1):
		    for j in range(1, m+1):
		        if s1[j-1]==s2[i-1]:
		            t[i][j]= 1+ t[i-1][j-1]
		        else:
		            t[i][j]= max(
		                t[i-1][j],
		                t[i][j-1]
		                )
	    lcs= t[-1][-1]
	    deletions= m-lcs
	    insertions= n-lcs
	    
	    return insertions+deletions
```
