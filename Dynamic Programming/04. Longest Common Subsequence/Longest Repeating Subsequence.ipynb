# [**Longest Repeating Subsequence**](https://practice.geeksforgeeks.org/problems/longest-repeating-subsequence2004/1)

```python
class Solution:
	def LongestRepeatingSubsequence(self, s):
		# Code here
		
		s1= s
		s2= s
		
		n= len(s)
		
		# initialization 
		t= [[0 for i in range(n+1)] for j in range(n+1)]
		
		# choice diagram
		for i in range(1, n+1):
		    for j in range(1, n+1):
		        if s1[i-1]==s2[j-1] and j!=i:
		            t[i][j]= 1+t[i-1][j-1]
		        else:
		            t[i][j]= max(
		                t[i-1][j],
		                t[i][j-1]
		                )
		return t[-1][-1]
```
