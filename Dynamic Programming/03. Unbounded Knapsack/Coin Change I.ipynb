# [**Coin Problem - I**](https://www.geeksforgeeks.org/problems/coin-change2448/1)

```python
#User function Template for python3

class Solution:
    def count(self, arr, n, Sum):
        # code here 
        t= [[0 for i in range(Sum+1)] for j in range(n+1)]
        
        # initialization
        for i in range(n+1):
            for j in range(Sum+1):
                if j==0:
                    t[i][j]=1 # when target sum is 0 , only one way to make that sum
                if n==0 and j>0:
                    t[i][j]=0 # no way to do
                    
        for i in range(1, n+1):
            for j in range(1, Sum+1):
                # if sum > arr[n-1]
                if arr[i-1]>j:
                    t[i][j]= t[i-1][j] # can't add coin
                else:
                    t[i][j]= t[i-1][j]+ t[i][j-arr[i-1]] # unbounded knapsack
                    
                    
        return t[-1][-1]
                    
                    
```
