# [**Target Size**](https://practice.geeksforgeeks.org/problems/target-sum-1626326450/1)

```python
class Solution:
    def count_target_sum(self, target_sum, arr, n):
        t= [[0 for i in range(target_sum+1)] for j in range(n+1)]

        for i in range(n+1):
            for j in range(target_sum+1):
                # initialization 
                if j==0 :
                    t[i][j]=1
                # if i==0 and j>0:
                #     t[i][j]=0

        for i in range(1, n+1):
            for j in range(1, target_sum+1):
                if arr[i-1]>j:
                    t[i][j]= t[i-1][j]
                else:
                    t[i][j]= (t[i-1][j] +  t[i-1][j-arr[i-1]])
        return t[-1][-1]% (int(1e9) + 7)
        






    def countPartitions(self, n, d, arr):
        # Code here
        # if sum(arr)%2==1:
        #     return 0
        s= (sum(arr)+d)
        arr.sort(reverse=True)
        if s%2==1 or abs(d)>sum(arr):
            return 0
        target_sum= s//2
        return self.count_target_sum(target_sum, arr, n)


s= Solution()
```

```python
n,d= 4, 3
arr= [5, 2, 6, 4]

s.countPartitions(n,d,arr)
```
