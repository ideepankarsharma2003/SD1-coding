# [**Coin Change I**](https://leetcode.com/problems/coin-change/submissions/)

```python
import sys
from typing import List

class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        n= len(coins)
        arr= coins
        s= amount

        t= [[sys.maxsize for i in range(s+1)]for j in range(n+1)]

        # initialization
        for i in range(n+1):
            for j in range(s+1):
                # if sum==0 and minimum number of coins required to do that
                if j==0:
                    t[i][j]=0
            
                # first row initialization is not needed as it is already initialized
                # n==0 and sum>0 : ----> impossible but -1 nhi kr skte because we have to do min
                # second row initialization
                if i==1 and j%arr[i-1]==0:
                    t[i][j]=j//arr[i-1]
                    


        for i in range(1, n+1):
            for j in range(1, s+1):
                if j<arr[i-1]:
                    t[i][j]= t[i-1][j]
                else:
                    t[i][j]= min(
                        t[i-1][j],          # dont consider
                        1+t[i][j-arr[i-1]]  # consider it, unbounded knapsack
                    )

        # for i in t:
        #     print(i)
        if  t[-1][-1]==sys.maxsize:
            return -1
        return t[-1][-1]


        
        
```
