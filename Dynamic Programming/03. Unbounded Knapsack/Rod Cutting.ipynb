# [**Rod Cutting Problem**](https://practice.geeksforgeeks.org/problems/rod-cutting0840/1)

```python
# Time limit Exceeded
class Solution:
    def cutRod(self, price, n):
        #code here
        # length= [i for i in range(1, n+1)] # length of pieces of rod
        t= [[0 for i in range(n+1)] for j in range(n+1)]
        
        for i in range(1, n+1):
            # i ----> length of piece of the rod
            for j in range(1, n+1):
                # processed
                # if length[i-1]>j:
                if i>j: # length of piece is greater than the actual length of the rod
                    t[i][j]=  t[i-1][j]# not select
                else:
                # choice
                    t[i][j]= max(
                            t[i-1][j],# not select
                            # price[i-1]+ t[i][j-length[i-1]] # select
                            price[i-1]+ t[i][j-i] # select
                    )
                    
                
        return t[-1][-1]
```

```python
#User function Template for python3

class Solution:
    def cutRod(self, price, n):
        #code here
        t = [0] * (n+1)  # space complexity O(1)
        for i in range(1, n+1):
            for j in range(i):
                t[i] = max(
                    t[i], # do not include
                    price[j] + t[i-1  -j] # include
                    
                )
        return t[n]
```
