# [**Longest Common Subsequence**](https://leetcode.com/problems/longest-common-subsequence/)

# Recursive

```python
class Solution:
    

    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m=len(text1)
        n=len(text2)
        
        def lcs(s1, s2, m, n):
            # base condition
            if n==0 or m==0:
                return 0

            if s1[m-1] == s2[n-1]:
                return 1+ lcs(s1, s2, m-1, n-1)
            else:
                return max(
                    lcs(s1, s2, m, n-1),
                    lcs(s1, s2, m-1, n)
                )
        return lcs(text1, text2, m, n)
```

# Memoization 

```python
class Solution:
    

    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m=len(text1)
        n=len(text2)

        self.t= [[-1 for i in range(m+1)] for j in range(n+1)]
        
        def lcs(s1, s2, m, n):
            # base condition
            if n==0 or m==0:
                return 0

            if self.t[n][m]!=-1:
                return self.t[n][m]

            if s1[m-1] == s2[n-1]:
                self.t[n][m]=  1+ lcs(s1, s2, m-1, n-1)
            else:
                self.t[n][m]= max(
                    lcs(s1, s2, m, n-1),
                    lcs(s1, s2, m-1, n)
                )
            return self.t[n][m]
        return lcs(text1, text2, m, n)
```
