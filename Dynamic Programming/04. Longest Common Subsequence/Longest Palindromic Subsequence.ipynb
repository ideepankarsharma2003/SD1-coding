# [**Longest Palindromic Subsequence**](https://leetcode.com/problems/longest-palindromic-subsequence/submissions/)<br>



```python
class Solution:
    def longestPalindromeSubseq(self, s: str) -> int:
        s1= s
        s2= s[::-1]
        n= len(s)
        m= n

        # LCS initialization
        t= [ [0 for j in range(m+1)] for i in range(n+1)]

        # LCS
        for i in range(1, n+1):
            for j in range(1, m+1):
                if s1[j-1]==s2[i-1]:
                    t[i][j]= 1+ t[i-1][j-1]
                else:
                    t[i][j]= max(
                        t[i-1][j],
                        t[i][j-1]
                    )

        return t[-1][-1]
            
```
