# [**Minimum number of deletions**](https://www.geeksforgeeks.org/problems/minimum-number-of-deletions4610/1)<br>



```python
#User function Template for python3
class Solution:
    def minDeletions(self, Str, n): 
        #code here
        
        s1= Str
        s2= Str[::-1]
        
        # initialization
        t= [[0 for i in range(n+1)] for j in range(n+1)]
        
        # LCS
        
        for i in range(1, n+1):
            for j in range(1, n+1):
                if s1[j-1]==s2[i-1]:
                    t[i][j]= 1+ t[i-1][j-1]
                else:
                    t[i][j]= max(
                        t[i-1][j],
                        t[i][j-1]
                        )
                        
        lps= t[-1][-1]
        
        min_dels= n-lps
        return min_dels
```
