# [**Is Subsequence**](https://leetcode.com/problems/is-subsequence/submissions/)



```python
class Solution:
    def isSubsequence(self, s1: str, s2: str) -> bool:
        # check if s1 is subsequence of s2
        n= len(s2)
        m= len(s1)

        # initialization
        t= [[0 for i in range(m+1)] for j in range(n+1)]

        # LCS
        for i in range(1, n+1):
            for j in range(1, m+1):
                if s1[j-1]==s2[i-1]:
                    t[i][j]= 1+ t[i-1][j-1]

                else:
                    t[i][j]= max(
                        t[i-1][j],
                        t[i][j-1]
                    )

        # is subsequence
        return t[-1][-1]==m
```
