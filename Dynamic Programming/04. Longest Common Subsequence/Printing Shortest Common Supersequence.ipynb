# [**Printing Shortest Common Supersequence**](https://leetcode.com/problems/shortest-common-supersequence/submissions/)

```python
class Solution:
    def shortestCommonSupersequence(self, str1: str, str2: str) -> str:
        s1 = str1
        m = len(s1)
        s2 = str2
        n = len(s2)

        # initialization
        t = [[0 for i in range(m+1)] for j in range(n+1)]

        # longest common subsequence
        for i in range(1, n+1):
            for j in range(1, m+1):
                if s2[i-1] == s1[j-1]:
                    t[i][j] = 1 + t[i-1][j-1]
                else:
                    t[i][j] = max(
                        t[i-1][j],
                        t[i][j-1]
                    )

        print(m+n-t[-1][-1])

        # shortest common supersequence
        i = n
        j = m

        ans = ""

        while i > 0 and j > 0:
            if s2[i-1] == s1[j-1]:
                ans += s2[i-1]
                i -= 1
                j -= 1
            else:
                if t[i-1][j] > t[i][j-1]:
                    ans += s2[i-1]
                    i -= 1
                else:
                    ans += s1[j-1]
                    j -= 1

        while i > 0:
            ans += s2[i-1]
            i -= 1
        while j > 0:
            ans += s1[j-1]
            j -= 1

        return ans[::-1]
```
